# KNN IMPLEMENTATION WITH BUILTIN DATASET 

IRIS which is a multiclass classification

```python
from sklearn.datasets import load_iris
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import random
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, f1_score,accuracy_score,confusion_matrix
```

```python
dataset = load_iris()
print(dataset.keys())
random.seed(10)
```

```python
dataset.data.shape
```

```python
independant_variables = pd.DataFrame(dataset.data,columns=dataset.feature_names)
independant_variables.head()
```

```python
dataset.target
```

```python
dataset.target_names
```

# Finding multicollinearity (correlation between independant variables)

```python
sns.pairplot(independant_variables)
plt.show()
```

# Train Test Split of data.

```python
train_x, test_x, train_y, test_y = train_test_split(independant_variables, dataset.target)
print(train_x[:5])
```

# Modelling

```python
classifier = KNeighborsClassifier()
model = classifier.fit(train_x, train_y)
```

# 
Predicting on the test set

```python
preds = model.predict(test_x)
print(preds[:10])
```

```python
print(test_y[:10])
```

# Error metrics

```python
print(accuracy_score(y_true=test_y, y_pred=preds))
```

```python
out =(confusion_matrix(y_true=test_y, y_pred=preds))
```
